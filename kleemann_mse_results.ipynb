```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
speed_up_mse = [0.0001530671725049615,
                0.00012985985085833818, 
                0.0001369089586660266,
                0.00043566991449348345]

noise_up_mse = [32.84317398071289,
                32.884490966796875, 
                32.58635330200195,
                186.33457889817868]

pressure_up_mse = [3.6679437160491943,
                   3.7938895225524902, 
                   3.7087481021881104,
                   3.686643904912635]

speed_down_mse = [0.0001374954590573907,
                  0.00012482416059356183, 
                  0.00011177156557096168,
                  0.0009505857370962825]

noise_down_mse = [23.53449249267578,
                  26.1539306640625, 
                  23.26426887512207,
                  23.473212404876147]

pressure_down_mse = [2.8346521854400635,
                    3.0737032890319824, 
                    2.8366451263427734,
                    2.1495999825438963]
```

```python
features = ["SPEED UP", "NOISE UP", "PRESSURE UP", "SPEED DOWN", "NOISE DOWN", "PRESSURE DOWN"]
archs = ["NN", "CNN", "LSTM", "ARIMA"]
mses = [speed_up_mse, noise_up_mse, pressure_up_mse, speed_down_mse, noise_down_mse, pressure_down_mse]
```

```python
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(7, 7))

for i in range(int(len(features) / 2)):
    feature_a, feature_b = features[i], features[i+3]
    
    # plot time signal:
    axs[i, 0].set_title(f"{feature_a}")
    axs[i, 0].bar(archs,mses[i])
    axs[i, 0].grid(axis='y')

    # plot different spectrum types:
    axs[i, 1].set_title(f"{feature_b}")
    axs[i, 1].bar(archs,mses[i+3])
    axs[i, 1].grid(axis='y')

fig.tight_layout()
plt.savefig("mses_results.png")
plt.show()
```
